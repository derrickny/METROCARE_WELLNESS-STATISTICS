### TOTAL NUMBER OF CLIENTS SEEN

```python
# Data to plot
labels = 'MALE', 'FEMALE'
sizes = [10, 9]
total = sum(sizes)

# Specify colors
colors = ['blue', 'lightgreen']

# Create a new figure with a specific size (width, height)
fig = plt.figure(figsize=(6, 6))

# Plot
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Add legend
legend_labels = [f'{label} ({size} out of {total})' for label, size in zip(labels, sizes)]
plt.legend(legend_labels, title="Total Clients Seen", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

# Save the figure as a PNG file
fig.savefig('pie_chart_Total_clients.png', bbox_inches='tight')

plt.show()
```
